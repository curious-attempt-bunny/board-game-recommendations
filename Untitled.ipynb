```python
import pandas as pd
import numpy as np
```

```python
ratings = pd.read_csv('game_ratings.csv.gz', encoding='utf-8', compression='gzip', index_col=0)
```

```python
ratings.drop(ratings.columns[ ratings.notnull().sum() < 50 ], axis=1, inplace=True)
```

```python
ratings.head()
```

```python
mean = ratings.mean(axis=0)
pd.DataFrame(mean).head()
```

```python
normalized = ratings.sub(mean, axis=1)
normalized.head()
```

```python
import sys

print('=' * normalized.index.shape[0])

sim = pd.DataFrame()

def similarity(i, j):
    filter = normalized.loc[i].notnull() & normalized.loc[j].notnull()
    
    if filter.sum() == 0:
        return 0
    
#     print '%d rows in common' % filter.sum()
    
    nom = (normalized.loc[i]*normalized.loc[j]).sum()
    
#     print 'nominator %g' % nom
    
#     print 'sum of ratings is %g' % (normalized.loc[i][filter].sum())
    denom = ((normalized.loc[i][filter]**2).sum()**0.5)*((normalized.loc[j][filter]**2).sum()**0.5)
    
#     print 'denominator %g' % denom
    return nom/denom

for i_index, i in enumerate(normalized.index.values):
    for j_index, j in enumerate(normalized.index.values):
#         print '%d,%d (%d, %d)' % (i_index, j_index, i, j)
        if (i_index <= j_index):
            sim.loc[i,j] = similarity(i,j)
        else:
            sim.loc[i,j] = sim.loc[j,i]

    sys.stdout.write('.')
    sys.stdout.flush()

# sim(175223,178900)
```
